Praktikum 01 - Bubble Sort

---



```python
def bubble_sort(data):
    n = len(data)
    for i in range(n):
        for j in range(0, n - i - 1):
            if data[j] > data[j + 1]:
                data[j], data[j + 1] = data[j + 1], data[j]
    return data

# Uji coba
angka = [64, 34, 25, 12, 22, 11, 90]
print("Data sebelum diurutkan:", angka)
print("Data setelah Bubble Sort:", bubble_sort(angka))
```

Praktikum 02 – Selection Sort

---



```python
def selection_sort(data):
    n = len(data)
    for i in range(n):
        min_index = i
        for j in range(i + 1, n):
            if data[j] < data[min_index]:
                min_index = j
        data[i], data[min_index] = data[min_index], data[i]
    return data

# Uji coba
angka = [64, 25, 12, 22, 11]
print("Data sebelum diurutkan:", angka)
print("Data setelah Selection Sort:", selection_sort(angka))
```

Praktikum 03 – Insertion Sort

---



```python
def insertion_sort(data):
    n = len(data)
    for i in range(1, n):
        key = data[i]
        j = i - 1
        while j >= 0 and data[j] > key:
            data[j + 1] = data[j]
            j -= 1
        data[j + 1] = key
    return data

# Uji coba
angka = [12, 11, 13, 5, 6]
print("Data sebelum diurutkan:", angka)
print("Data setelah Insertion Sort:", insertion_sort(angka))
```

Praktikum 04 – Studi Kasus: Mengurutkan Nilai Mahasiswa

---



```python
def bubble_sort_nilai(data):
    n = len(data)
    for i in range(n):
        for j in range(0, n - i - 1):
            if data[j][1] < data[j + 1][1]:
                data[j], data[j + 1] = data[j + 1], data[j]
    return data

# Data: List berisi Tuple (Nama, Nilai)
mahasiswa = [("Andi", 80), ("Budi", 70), ("Cici", 90), ("Doni", 85)]
print("Data sebelum diurutkan:", mahasiswa)
print("Data setelah diurutkan (desc):", bubble_sort_nilai(mahasiswa))
```

Praktikum 06 – Sorting Data Barang (Nama & Harga)

---



```python
def bubble_sort_key(data, key_index=0, reverse=False):
    n = len(data)
    for i in range(n):
        for j in range(0, n - i - 1):
            if reverse:
                if data[j][key_index] < data[j + 1][key_index]:
                    data[j], data[j + 1] = data[j + 1], data[j]
            else:
                if data[j][key_index] > data[j + 1][key_index]:
                    data[j], data[j + 1] = data[j + 1], data[j]
    return data

# Data Barang (Nama, Harga)
barang = [("Buku", 15000), ("Pensil", 5000), ("Tas", 75000), ("Pulpen", 10000)]
print("Urut berdasarkan nama (asc):", bubble_sort_key(barang.copy(), 0))
print("Urut berdasarkan harga (desc):", bubble_sort_key(barang.copy(), 1, reverse=True))
```

3. Eksperimen Bubble Sort

---



```python
print("Ahmad Fajril Falah")
print("Eksperimen Bubble Sort (Sorting)\n")

#Ascending - kodingan dari praktikum
def bubble_sort_ascending(data):
    n = len(data)
    arr = data.copy()
    for i in range(n):
        for j in range(0, n - i - 1):
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
    return arr

#Descending - hasil modifikasi, hanya diubah tanda "<" dan ">"
def bubble_sort_descending(data):
    n = len(data)
    arr = data.copy()
    for i in range(n):
        for j in range(0, n - i - 1):
            if arr[j] < arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
    return arr

# Uji coba
angka = [64, 34, 25, 12, 22, 11, 90]

print("Data Awal             :", angka)
print("-" * 50)
print("Hasil Ascending (>)   :", bubble_sort_ascending(angka))
print("Hasil Descending (<)  :", bubble_sort_descending(angka))
```

4.	Eksperimen Selection Sort

---



```python
print("Ahmad Fajril Falah")
print("4.Eksperimen Selection Sort\n  (Mencari elemen terbesar dan menempatkannya di posisi akhir)\n")

def selection_sort_max(data):
    n = len(data)
    for i in range(n - 1, 0, -1):
        max_index = 0
        for j in range(1, i + 1):
            if data[j] > data[max_index]:
                max_index = j
        data[i], data[max_index] = data[max_index], data[i]

    return data

# Uji coba
angka = [64, 25, 12, 22, 11]
print(f"Data Awal        : {angka}")
print(f"Hasil Modifikasi : {selection_sort_max(angka)}")
```

5.	Eksperimen Insertion Sort

---



```python
print("Ahmad Fajril Falah")
print("5.Eksperimen Insertion Sort\n")

def insertion_sort_nama(data):
    n = len(data)
    for i in range(1, n):
        key = data[i]
        j = i - 1
        while j >= 0 and data[j] > key:
            data[j + 1] = data[j]
            j -= 1
        data[j + 1] = key

    return data

# Uji coba
mahasiswa = ["Fajril", "Alif", "Daffa", "Irfan", "Nafi"]
print("Daftar Awal      :", mahasiswa)
print("Urut Alfabetis   :", insertion_sort_nama(mahasiswa))
```

7.	Analisis Kompleksitas

---



```python
print("Ahmad Fajril Falah")
print("7.Analisis Kompleksitas\n")

import random

# bubble sort
def count_bubble_sort(data):
    arr = data.copy()
    count = 0
    n = len(arr)
    for i in range(n):
        for j in range(0, n - i - 1):
            count += 1
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
    return count

# selection sort
def count_selection_sort(data):
    arr = data.copy()
    count = 0
    n = len(arr)
    for i in range(n):
        min_index = i
        for j in range(i + 1, n):
            count += 1
            if arr[j] < arr[min_index]:
                min_index = j
        arr[i], arr[min_index] = arr[min_index], arr[i]
    return count

# insertion sort
def count_insertion_sort(data):
    arr = data.copy()
    count = 0
    n = len(arr)
    for i in range(1, n):
        key = arr[i]
        j = i - 1

        while j >= 0:
            count += 1
            if arr[j] > key:
                arr[j + 1] = arr[j]
                j -= 1
            else:
                break

        arr[j + 1] = key
    return count

#TABEL ANALISIS
print(f"{'UKURAN':<10} | {'BUBBLE':<10} | {'SELECTION':<10} | {'INSERTION':<10}")
print("-" * 50)

ukuran_data = [10, 50, 100]

for n in ukuran_data:
    data_awal = [random.randint(1, 1000) for _ in range(n)]

    c_bubble = count_bubble_sort(data_awal)
    c_select = count_selection_sort(data_awal)
    c_insert = count_insertion_sort(data_awal)

    print(f"{n:<10} | {c_bubble:<10} | {c_select:<10} | {c_insert:<10}")
```

```python
print("Ahmad Fajril Falah")
print("Proyek Mini: Manajemen Barang\n")

# INPUT DATA
daftar_barang = []
try:
    jumlah = int(input("Masukkan jumlah barang: "))
    for i in range(jumlah):
        print(f"Data ke-{i+1}")
        nama = input("  Nama Barang : ")
        harga = int(input("  Harga Barang: "))
        daftar_barang.append((nama, harga))

    # PILIHAN MENU SORTING
    print("\n--- Pilihan Pengurutan ---")
    print("1. Nama (A-Z) Ascending")
    print("2. Nama (Z-A) Descending")
    print("3. Harga (Murah-Mahal) Ascending")
    print("4. Harga (Mahal-Murah) Descending")

    pilihan = int(input("Pilih metode (1-4): "))

    # LOGIKA SORTING
    if pilihan == 1:
        daftar_barang.sort(key=lambda x: x[0])                 # Nama, Naik
        ket = "Nama (A-Z)"
    elif pilihan == 2:
        daftar_barang.sort(key=lambda x: x[0], reverse=True)   # Nama, Turun
        ket = "Nama (Z-A)"
    elif pilihan == 3:
        daftar_barang.sort(key=lambda x: x[1])                 # Harga, Naik
        ket = "Harga (Termurah)"
    elif pilihan == 4:
        daftar_barang.sort(key=lambda x: x[1], reverse=True)   # Harga, Turun
        ket = "Harga (Termahal)"
    else:
        print("Pilihan tidak valid! Data tidak diurutkan.")
        ket = "Tidak Diurutkan"

    # HASILNYA
    print(f"\n--- Hasil Urut Berdasarkan {ket} ---")
    print(f"{'No':<4} | {'Nama Barang':<15} | {'Harga (Rp)':<15}")
    print("-" * 40)
    for i, item in enumerate(daftar_barang):
        print(f"{i+1:<4} | {item[0]:<15} | {item[1]:<15}")

except ValueError:
    print("\nError: Masukkan angka yang valid untuk harga/jumlah!")
```
